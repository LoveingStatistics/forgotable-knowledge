## 层次索引

- series/dataframe:
    - 将index参数使用列表的列表进行设置
    - 可以使用loc进行索引，同时可以内外层索引
    - data.unstack(level=0/1,fill_values=''),level可以传递索引的name
    - data.stack()
    - data.swaplevel(index1,index2,axis=0/1)
    - data.sort_index(level=0/1)
    - data.sum(axis=0/1,level=多重索引中的一层)
    - 对于dataframe，也可以进行列的多层索引操作
        - data.set_index(['a','b'],replace=False)
        - data2.reset_index()进行反向操作

Series

```python
data = pd.Series(np.random.randn(9),
                 index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],
                        [1, 2, 3, 1, 3, 1, 2, 2, 3]]) #使用列表的列表进行创建
data
```

### 重新排序

```python
#交换双重索引的内外顺序
frame.swaplevel('key1','key2',axis=0)

#对内索引排序
frame.sort_index(level=1)
#交换索引和重新排序
frame.swaplevel(0, 1).sort_index(level=0)
```

## 合并数据集

- pd.merge()和SQL中的join方法类似，根据指定的列进行拼接
    - 多对一和多对多，通过指定on参数设定连接的键
    - 通过how参数指定连接方法：outer,inner.left,right
    - 如果是多重键连接的话，列名称使用列表进行指定
    - suffixes指定重复列的后缀
    - right/left_index=True/False指定是否使用index作为连接键
- pd.concat()沿着某个轴进行合并
    - 通过list（列表）传递合并的数据集，axis=0/1指定列合并还是行合并
    - join指定连接的方法，内外连接
    - keys指定层次索引说明数据来自哪里
    - igorne_index=True重置合并后的索引
- pd.combine_first()进行填补漏洞的合并
    - 用另一个数据的值填补前面的空洞
    - 行列的索引值是两个数据集的合并之后的结果，有可能产生新的缺失值

### pd.merge()

#### 多对一

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                    'data2': range(3)})

pd.merge(df1,df2, on='key',how='right')
'''连接方法：外连接'''
pd.merge(df1, df2, how='outer')
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1': range(7)})
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                    'data2': range(3)})
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

#### 多对多

对匹配到的索引值进行笛卡尔积连接

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],
                    'data1': range(6)})
df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                    'data2': range(5)})
df1
df2
pd.merge(df1,df2,how='outer',on='key')
```

```python
#多重主键,使用列表进行指定
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                     'key2': ['one', 'two', 'one'],
                     'lval': [1, 2, 3]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                      'key2': ['one', 'one', 'one', 'two'],
                      'rval': [4, 5, 6, 7]})
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
'''重复的列名称的处理：添加后缀'''
pd.merge(left, right, on='key1')
pd.merge(left,right,on='key1',suffixes=('_left','_right'))
```

#### 使用index作为键进行合并

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                      'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
left1
right1
```

```python
'''内连接'''
pd.merge(left1, right1, left_on='key', right_index=True)
'''外连接'''
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

多重索引下如何使用

```python
lefth = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio',
                               'Nevada', 'Nevada'],
                      'key2': [2000, 2001, 2002, 2001, 2002],
                      'data': np.arange(5.)})
righth = pd.DataFrame(np.arange(12).reshape((6, 2)),
                      index=[['Nevada', 'Nevada', 'Ohio', 'Ohio',
                              'Ohio', 'Ohio'],
                             [2001, 2000, 2000, 2000, 2001, 2002]],
                      columns=['event1', 'event2'])
lefth
righth
```

### pd.concat()

#### index不一样的时候，会如何连接（可以自定义）

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f','g'])
pd.concat([s1,s2,s3]) #rbind

pd.concat([s1, s2, s3], axis=1,sort=False) #cbind,默认进行外连接
```

```python
s4 = pd.concat([s1, s3])
s4
s1
pd.concat([s1, s4], axis=1, join='inner') #cbind,同时指定为内连接
```

```python
pd.concat([s1, s4], axis=1,sort=False).reindex(['a', 'c', 'b', 'e']) #还可以指定显示哪些index
```

#### 设置层次索引来表明属于哪个数据集

```python
#rbind，就是设置层次索引
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three'])
result
result.unstack() #和下面的结果一样
```

```python
#cbind,就是指定列名称
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three'],sort=False)
```

```python
'''数据框的结果也类似，不过cbind就是层次列名称了'''
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'],
                   columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'],
                   columns=['three', 'four'])
df1
df2
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'],sort=False,names=['outer','inner'])
```

### data1.combine_first(data2)：‘补漏洞’

- 用另一个数据的值填补前面的空洞

pandas的用法：combine_first

```python
a.combine_first(b)
```

```python
df1.combine_first(df2)
```

```python
'''转化为长数据'''
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter,
                         name='date') #时间型的索引
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item') #普通索引
data = data.reindex(columns=columns) #相当于data.loc[:,[columns]]
data.index = periods.to_timestamp('D', 'end') #将时间型索引进行时间戳转化
data.head(5)
ldata = data.stack().reset_index().rename(columns={0: 'value'}) #重命名，使用rename并结合字典
ldata.head(5)
```

```python
'''长变成宽'''
pivoted = ldata.pivot('date', 'item', 'value')
pivoted
```

```python
ldata['value2'] = np.random.randn(len(ldata))
pivoted = ldata.pivot('date', 'item')
pivoted.head(5)
#调用外层索引
pivoted['value'].head()
```

# 数据的聚合操作

聚合操作可以拆分成三部分，如下图所示：

## 原理介绍

- 聚合的根据可以是数据框中的一列或者多列
- 也可以是和数据框等长的array

### 基本操作，通过列名称group

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

依据一列的值进行聚合

```python
grouped = df['data1'].groupby(df['key1'])
grouped #是一个分组对象，只分组不操作
grouped.mean()
```

聚合依据是多个列

```python
means = df['data1'].groupby([df['key1'], df['key2']]).mean()
means
means.unstack()
```

可以是等长的array

```python
states = np.array(['Ohio', 'California', 'California', 'Ohio', 'Ohio'])
years = np.array([2005, 2005, 2006, 2005, 2006])
df['data1'].groupby([states, years]).mean()
```

```python
df.groupby(['key1', 'key2']).size()
```

### 对聚合对象进行迭代

```python
for name, group in df.groupby('key1'):
    print(name)
    print(group)
```

```python
for keys, group in df.groupby(['key1', 'key2']):
    print(keys)
    print(group)
```

```python
dict(list(df.groupby('key1'))) #使用字典调用
```

```python
'''列聚合'''
grouped = df.groupby(df.dtypes, axis=1)
for types, groups in grouped:
    print(types)
    print(groups)
```

### 使用字典进行group

```python
'''数据框形式的抽取还是Series的抽取'''
df.groupby('key1')[['data1']]
df.groupby('key1')['data1']
```

```python
people = pd.DataFrame(np.random.randn(5, 5),
                      columns=['a', 'b', 'c', 'd', 'e'],
                      index=['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
people.iloc[2:3, [1, 2]] = np.nan
people
```
