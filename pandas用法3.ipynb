# 数据的聚合操作

## 聚合

- grouped.mean().count().max()
    - 或者自定义函数
- grouped.agg([函数方法（聚合操作）])
- grouped.agg([('play','mean'),('yeah',func)])
- grouped.agg({'column_name':'mean','column_name2':[方法]})
- df.groupby('key1',as_index=False).agg()
    - 或者df.groupby('key1').agg().reset_index()

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

```python
grouped = df.groupby('key1')
grouped['data1'].quantile(0.9)
```

```python
'''自定义函数进行汇总'''
def peak_to_peak(arr):
    return arr.max() - arr.min()

grouped.agg(peak_to_peak)
```

```python
grouped['data1'].describe() #甚至可以这样
```

```python
tips = pd.read_csv('examples/tips.csv')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
grouped = tips.groupby(['day', 'smoker'])
grouped_pct = grouped['tip_pct']
grouped_pct.agg('mean')
```

```python
'''使用列表传入多个方法'''
grouped_pct.agg(['mean', 'std', peak_to_peak])
```

```python
'''为使用的方法重命名显示'''
grouped_pct.agg([('foo', 'mean'), ('bar', np.std)])
```

```python
functions = ['count', 'mean', 'max']
grouped['tip_pct','total_bill'].agg(functions)
```

```python
ftuples = [('Durchschnitt', 'mean'), ('Abweichung', np.var)]
grouped['tip_pct', 'total_bill'].agg(ftuples)
```

```python
'''不同列使用不同的函数'''
grouped.agg({'tip' : np.max, 'size' : 'sum'})
```

```python
grouped.agg({'tip_pct' : ['min', 'max', 'mean', 'std'],
             'size' : 'sum'})
```

```python
'''重置index'''
grouped = tips.groupby(['day', 'smoker'])
grouped.agg({'tip_pct' : ['min', 'max', 'mean', 'std'],
             'size' : 'sum'}).reset_index()

grouped = tips.groupby(['day', 'smoker'],as_index=False)
grouped.agg({'tip_pct' : ['min', 'max', 'mean', 'std'],
             'size' : 'sum'})
```

## apply:另一种聚合操作

- np.average(data,weights=)进行加权平均
- 对每个group实行apply中的操作之后再将它们合起来
- df.groupby('key',group_keys=False).sum()不显示组的名称
- group.name可以调用组名称然后针对组名称实行不同的操作
- pd.read_csv('examples/stock_px_2.csv', parse_dates=True,index_col=0)parse_dates=True将字符型数据转化为日期型
- data.pct.change()计算百分比变化率

```python
tips = pd.read_csv('examples/tips.csv')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
def top(df,n=5,column='tip_pct'):
    return df.sort_values(by=column)[-n:]

top(tips,n=6)
```

```python
#分组进行排序
tips.groupby('smoker').apply(top)
```

```python
tips.groupby('smoker').apply(top,n=6,column='size')
```

```python
tips.groupby('smoker', group_keys=False).apply(top)
```

### many examples

#### 按照分箱进行不同操作，体会apply和agg的区别

```python
frame = pd.DataFrame({'data1': np.random.randn(1000),
                      'data2': np.random.randn(1000)})
quartiles = pd.cut(frame.data1, 4)
quartiles[:10]
```

```python
'''agg和apply的区别'''
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}

grouped = frame.data2.groupby(quartiles)
grouped.apply(get_stats)
grouped.agg(get_stats)
```

```python
grouping = pd.qcut(frame.data1, 10,labels=False)
grouping.head(3)
grouped = frame.data2.groupby(grouping)
grouped.apply(get_stats).unstack()
```

#### 分组填充缺失值

```python
s = pd.Series(np.random.randn(6))
s[::2] = np.nan
s
```

```python
s.fillna(s.mean())
```

```python
states = ['Ohio', 'New York', 'Vermont', 'Florida',
          'Oregon', 'Nevada', 'California', 'Idaho']
group_key = ['East'] * 4 + ['West'] * 4
data = pd.Series(np.random.randn(8), index=states)
data[['Vermont', 'Nevada', 'Idaho']] = np.nan
data
```

```python
f = lambda x: x.fillna(x.mean())
data.groupby(group_key).apply(f)
```

```python
'''apply中使用name可以调用组名称，然后根据组名称的不同使用定义在字典中对应的操作'''
fill_values = {'East': 0.5, 'West': -1}
fill_func = lambda g: g.fillna(fill_values[g.name])
data.groupby(group_key).apply(fill_func)
```

#### 分组抽样和随机

```python
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10] * 3) * 4
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
cards = []
for suit in suits:
    cards.extend(str(num) + suit for num in base_names)

deck = pd.Series(card_val, index=cards)
deck.head()
```

```python
def draw(deck, n=5):
    return deck.sample(n)
draw(deck,3)
```

```python
get_suit = lambda card: card[-1] 
deck.groupby(get_suit).apply(draw, n=2)

deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

#### 分组描述性统计：加权平均、相关系数

```python
df = pd.DataFrame({'category': ['a', 'a', 'a', 'a',
                                'b', 'b', 'b', 'b'],
                   'data': np.random.randn(8),
                   'weights': np.random.rand(8)})
df
```

```python
grouped = df.groupby('category')
get_wavg = lambda g: np.average(g['data'], weights=g['weights'])
grouped.apply(get_wavg)
```

```python
close_px = pd.read_csv('examples/stock_px_2.csv', parse_dates=True,
                       index_col=0)
close_px.info()
close_px.head()
```

```python
spx_corr = lambda x: x.corrwith(x['SPX']) #计算相关系数的函数
rets = close_px.pct_change().dropna()

'''按年份分组'''
get_year = lambda x: x.year
by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
by_year.apply(lambda x: x['AAPL'].corr(x['MSFT']))
```

#### 分组建模(回归分析)

```python
import statsmodels.api as sm #引入回归函数
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params

by_year.apply(regress,'SPX',['AAPL','MSFT','XOM'])
```

## 时区的处理

1. UTC格林维尔世界标准时间
    - pytz包可以处理时区问题
    - data.tz_localize('UTC')时区代码，指定当前时间是哪个时区的时间
    - data.tz_convert('China')转换时区显示：-2012-03-09 04:30:00-05:00-后面的-5:00表示比标准时间晚5个小时
    - 每年3月11日，纽约开始实行日光节约时间，所以时间会播快一个小时，使得和UTC的时间差少一个小时
2. index涉及时区时无需担心，因为会转化成UTC时间进行统一计算和匹配

## resampling:时序数据特殊的groupby,容易弄错，参数最好明确

1. data.resampling().方法
    1. 一个适用于股市的方法，data.resampling().ohlc()显示分组的开盘收盘，最高最低值
    2. 可以大致分为上采样和下采用
    3. downsampling：
        1. 可以把时间频率放慢，可以选择参数：closed决定左右区间那边闭合
        2. label显示左右端点的哪一个；
        3. loffset可以对端点值进行一些操作
    4. upsampling:
        1. 可以把时间频率加快。
        2. data.resampling().asfreq()能够显示出来结果
        3. data.resampling().ffill()进行向前填充缺失值
2. 以上都是timestamp的resampling方法
    1. 这里是period的resampling方法
    2. 下采样放慢区间时，closed参数好像没有用了
    3. 上采样的时候需要使用convention='start/end'方法来决定选取哪个细分区间，类似period.asfreq(how='end/start')

```python
rng = pd.date_range('2000-01-01', periods=100, freq='D')
ts = pd.Series(np.random.randn(len(rng)),index=rng)
ts
```

```python
ts.resample('M').mean()
ts.resample('M', kind='period').mean()
```

### downsampling:把时间频率减小

```python
rng = pd.date_range('2000-01-01', periods=12, freq='T')
ts = pd.Series(np.arange(12), index=rng)
ts
```

```python
'''对比closed不同的区别，区间代表label就不同'''
ts.resample('5min',closed='right').sum()
ts.resample('5min',closed='left').sum()
```

```python
'''右边闭区间的话，可以通过label来调整显示左端点还是右端点'''
ts.resample('5min', closed='right', label='right').sum()
```

```python
ts.resample('5min', closed='right', label='right',loffset='1s').sum()
```

```python
'''一键获取时间段内的开盘收盘，最高最低值，股票数据很好用'''
ts.resample('5min').ohlc()
```

### resampling with periods

```python
frame = pd.DataFrame(np.random.randn(24, 4),
                     index=pd.period_range('1-2000', '12-2001',
                                           freq='M'),
                     columns=['Colorado', 'Texas', 'New York', 'Ohio'])
frame.head()
```

```python
annual_frame = frame.resample('A-DEC').mean()
annual_frame
```

```python
'''上采样periods怎么采'''
annual_frame.resample('Q-DEC').ffill()
annual_frame.resample('Q-DEC',convention='end').ffill()
```

```python
'''something interesting，注意annual_frame的freq是A-DEC'''
annual_frame.resample('Q-MAR').ffill()
```

```python
'''当不成倍数的时候就可能会产生多的数据'''
annual_frame.resample('Q-JAN').ffill()
annual_frame.resample('Q-JAN',convention='end').ffill()
```
