## index和value原理

index和value的原理非常类似字典，可以通过index定位到对应的value，这个原理十分的关键

```python
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
obj1=pd.Series(sdata)
states = ['California', 'Ohio', 'Oregon', 'Texas']
obj2=pd.Series(sdata,index=states)
obj2 #pandas中的index和value很像字典这种数据结构
```

```python
'''Series最重要的一个功能是： 它在算术运算中会自动对齐不同索引的数据。'''
obj1+obj2 #对相同的index进行相加
```

```python
'''Series对象本身及其索引都有一个name属性， 该属性跟pandas其他的关键功能关系非常密切'''
obj2.name='population'
obj2.index.name='state'
obj2
```

```python
'''Series的索引可以通过赋值的方式就地修改'''
obj1.index=['a','b','c','d']
obj1
```

### reindex重新索引

pandas对象的一个重要方法是reindex， 其作用是创建一个适应新索引的新对象

```python
'''重新索引'''
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
obj.reindex(['a','b','c','d','e'],fill_value=0) #重新索引，并且填补新索引对应的缺失值
```

```python
obj2 = pd.Series(['blue', 'purple', 'yellow'], index=[0, 2, 4])
obj2.reindex(range(6),method='ffill') #向前填充索引对应的值
#还可以向后索引，method=‘bfill’
```

重新索引也可以对应列索引，这样就可以在创建完数据框之后进行列之间的顺序调整，类似excel的列拖动调整

```python
frame3 = pd.DataFrame(np.arange(9).reshape((3, 3)), index=['a', 'c', 'd'],
    columns=['Ohio', 'Texas', 'California'])

frame3.reindex(columns=['Texas','California']) 
#重新索引也可以对应列索引，这样就可以在创建完数据框之后进行列之间的顺序调整，类似excel的列拖动调整
```

```python
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)), columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)), columns=list('abcde'))
```

```python
df1
```

```python
df2
```

```python
df2.add(df1,fill_value=0) #在维度不相等的时候进行0填充，而不是返回一个nan值
```

```python
df1.reindex(columns=df2.columns, fill_value=0)
```

### 使用lxml解析XML源文件

#### 读取网页

- 使用lxm.objectify进行网页解析
- objectify.parse进行XML解析
  - parse在解析的时候需要从内存中解析，所以需要将本地文件open，或者将一个字符串（网页元素）StringIO【输入到内存】
- getroot获得根节点
- root.INDICATOR返回一个生成器，遍历XML文件的所有INDICATOR
  - 循环root.INDICATOR，对于每一个INDICATOR，使用getchildren()方法来获取其中的元素
  - 对于每一个child，child.tag调用标签，child.pyval调用值

XML格式是比HTML更为泛化的一种网页内容存储文件格式

```python
from lxml import objectify
```

XML文件示例，一个INDICATOR是一个单元，解析的时候可以一个一个单元的解析

<INDICATOR>
    <INDICATOR_SEQ>373889</INDICATOR_SEQ>
    <PARENT_SEQ></PARENT_SEQ>
    <AGENCY_NAME>Metro-North Railroad</AGENCY_NAME>
    <INDICATOR_NAME>Escalator Availability</INDICATOR_NAME>
    <DESCRIPTION>Percent of the time that escalators are operational
    systemwide. The availability rate is based on physical observations performed
    the morning of regular business days only. This is a new indicator the agency
    began reporting in 2009.</DESCRIPTION>
    <PERIOD_YEAR>2011</PERIOD_YEAR>
    <PERIOD_MONTH>12</PERIOD_MONTH>
    <CATEGORY>Service Indicators</CATEGORY>
    <FREQUENCY>M</FREQUENCY>
    <DESIRED_CHANGE>U</DESIRED_CHANGE>
    <INDICATOR_UNIT>%</INDICATOR_UNIT>
    <DECIMAL_PLACES>1</DECIMAL_PLACES>
    <YTD_TARGET>97.00</YTD_TARGET>
    <YTD_ACTUAL></YTD_ACTUAL>
    <MONTHLY_TARGET>97.00</MONTHLY_TARGET>
    <MONTHLY_ACTUAL></MONTHLY_ACTUAL>
</INDICATOR>

```python
path = 'examples/datasets/mta_perf/Performance_MNR.xml'
parsed = objectify.parse(open(path)) #解析XML
root = parsed.getroot() #获得根节点
```

```python
data = []
skip_fields = ['PARENT_SEQ', 'INDICATOR_SEQ','DESIRED_CHANGE', 'DECIMAL_PLACES'] #定义不需要的元素

for elt in root.INDICATOR: #遍历所有的单元
    el_data = {}
    for child in elt.getchildren(): #遍历单元中的所有元素
        if child.tag in skip_fields:
            continue
        el_data[child.tag] = child.pyval
    data.append(el_data)
```

      - 其中table虽然比较慢，但是支持查询:store.select(,where=[])

### pd.HDFStore

```python
frame = pd.DataFrame({'a': np.random.randn(100)})
store = pd.HDFStore('mydata.h5')

store['obj1'] = frame
store['obj1_col'] = frame['a']
store
```

#### 如果一个样本属于多个类别，那么如何转换呢？

电影数据集

首先，读取数据

```python
mnames = ['movie_id', 'title', 'genres']
#read_table读取数据
movies = pd.read_table('examples/datasets/movielens/movies.dat', sep='::',
                       header=None, names=mnames,engine='python')

movies[:10]
```

接下来获得不同类别的数量

注意extend和append的区别

```python
all_genre = []
for x in movies['genres']:
    all_genre.extend(x.split('|'))

genres = pd.unique(all_genre)
```

```python
#共有18种类型的电影
len(genres)
```

```python
zero_matrix = np.zeros((len(movies), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)

###循环遍历整个样本集，然后对每个样本的种类首先进行分词，然后对应位置放1
gen = movies.genres[0]
gen.split('|')

dummies.columns.get_indexer(gen.split('|'))
```

遍历进行设置

```python
for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[:3,]
```

## 字符串处理

- 注意：在字符串输入的时候，为了不把\当作转义字符，可以在字符前面添加r'C:\Users\89592\Desktop'
- 字符串基本操作
    - data.split(',') 字符串分离
    - x.strip()去掉字符串前后的空白，不能去掉中间的空白
    - ';'.join(data)添加分隔符
    - 定位字符串部分内容的位置
        - in 判断是否存在
        - data.index() 子字串第一次出现的位置，没有的话会报错
        - data.find() 如果没有，会返回-1，其他的和index一样
- 正则化表达式
    - re包：用来比对，查找和替换字符串内容
    - re.split('\s+',data) 根据表达式分割
    - regex = re.compile('\s+');regex.split(data),这样可以节约内存空间
    - re.compile('\s+',flags=re.IGNORECASE)是指忽略大小写
        - '\s'：s是space的缩写，所以指代了空格，'+'就代表大于等于一个空格
        - r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}' #转义字符使得.就是.    这个正则表达式第一项，是允许[]中的元素出现一次，'+'代表前面的表达式可以出现很多次;{}代表前面的表达式至少出现2次，至多4次
    - regex.findall()寻找符合正则表达式的部分
    - regex.search()返回字符串中第一个出现的匹配项
        - search返回的是一个特殊的对象，如果想要查找，需要使用data\[rgex.search().start():rgex.search().start()\]
    - regex.match()返回字符串的开头是否符合正则表达式
    - regex.sub('AAA',data) 把符合的表达式替换为'AAA'
    - 正则化表达式中使用（）进行分割表达
        - regex.match(data).groups(),分割
        - regex.findall(data)
        - regex.sub(r'Username: \1, Domain: \2, Suffix: \3', text):\1\2\3进行指定形式的分割
- pandas中向量化处理字符串
     - 如果直接使用apply和re函数，可能无法处理缺失值
     - pandas内置了str方法可以跳过缺失值同时向量化处理每个元素
         - data.str.contains('china')
         - data.str.findall(pattern,flags=re.IGNORECASE)
         - data.str.match(pattern,flags=re.IGNORECASE)

### 正则化表达式

find

```python
import re
#按照空格分离
text = "foo bar\t baz \tqux"
re.split('\s+', text)
```

```python
regex = re.compile('\s+')
regex.split(text)
```

```python
#在字符串中查找正则化表达式
regex.findall(text)
```

```python
'''一个略微复杂的例子'''
text = """Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}' #转义字符使得.就是.
#这个正则表达式第一项，是允许[]中的元素出现一次，'+'代表前面的表达式可以出现很多次;{}代表前面的表达式至少出现2次，至多4次
regex = re.compile(pattern, flags=re.IGNORECASE)

regex.findall(text)
```

```python
#search()进行第一次检索
m = regex.search(text)

m #search是一个对象
text[m.start():m.end()] #将search的内容提取出来
```

```python
#match返回字符串的开头是否符合正则表达式，不符合返回None
print(regex.match(text))
```

substitute

```python
print(regex.sub('REDACTED', text))
```

split：把邮箱分成前缀、网域名称、和最后的后缀

```python
pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})' #注意这里的正则表达式的每一个小部分都用（）括了起来
regex = re.compile(pattern, flags=re.IGNORECASE)

m = regex.match('wesm@bright.net')
m.groups()
```

```python
regex.findall(text)
```

```python
#sub（使用\1\2\3进行指定的添加）
print(regex.sub(r'Username: \1, Domain: \2, Suffix: \3', text))
```
